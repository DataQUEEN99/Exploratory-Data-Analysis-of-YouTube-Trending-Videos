```python
import pandas as pd

df = pd.read_csv("USvideos[1].csv")
df.head()
```

```python
import json

# Load JSON
with open("US_category_id[1].json") as f:
    categories = json.load(f)

# Make dictionary {id: category_name}
category_dict = {}
for item in categories['items']:
    category_dict[int(item['id'])] = item['snippet']['title']

# Add new column to dataframe
df['category_name'] = df['category_id'].map(category_dict)

# Show first few rows with category name
df[['title', 'category_id', 'category_name']].head()
```

```python
#univariate data
#. Most Common Categories in Trending Videos
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.countplot(y='category_name', data=df, order=df['category_name'].value_counts().index, palette="viridis")
plt.title("Most Common Categories in Trending Videos")
plt.xlabel("Number of Videos")
plt.ylabel("Category")
plt.show()


```

```python
#Distribution of Views
plt.figure(figsize=(8,5))
sns.histplot(df['views'], bins=50, kde=True, color="blue")
plt.title("Distribution of Views")
plt.xlabel("Views")
plt.ylabel("Frequency")
plt.show()
```

```python
#top 10 most viewed videos
top10_views = df.sort_values(by='views', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x='views', y='title', data=top10_views, palette="magma")
plt.title("Top 10 Most Viewed Trending Videos")
plt.xlabel("Views")
plt.ylabel("Video Title")
plt.show()
```

```python
#Bivariate Analysis
#1 Average Views by Category
avg_views = df.groupby('category_name')['views'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
avg_views.plot(kind='bar', color='skyblue')
plt.title("Average Views by Category")
plt.ylabel("Average Views")
plt.xlabel("Category")
plt.show()
```

```python
## views vs comment count
plt.figure(figsize=(8,6))
sns.scatterplot(x='views', y='comment_count', data=df, alpha=0.4, color="green")
plt.title("Views vs Comments")
plt.xlabel("Views")
plt.ylabel("Comment Count")
plt.show()
```

```python
##correlation heatmap
plt.figure(figsize=(7,5))
sns.heatmap(df[['views','likes','dislikes','comment_count']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#engagement analysis
##add engagement column
# Engagement = Likes + Comments
df['engagement'] = df['likes'] + df['comment_count']

# Show first few rows
df[['title','category_name','views','likes','comment_count','engagement']].head()
```

```python
##Average Engagement by Category
#Which categories create the most active audience?
avg_engagement = df.groupby('category_name')['engagement'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
avg_engagement.plot(kind='bar', color='coral')
plt.title("Average Engagement by Category")
plt.ylabel("Average Engagement (Likes + Comments)")
plt.xlabel("Category")
plt.show()
```

```python
##Engagement Rate (Engagement ÷ Views)

##Sometimes a video may have millions of views but low interaction.
#So let’s calculate engagement rate
df['engagement_rate'] = (df['engagement'] / df['views']) * 100

avg_rate = df.groupby('category_name')['engagement_rate'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
avg_rate.plot(kind='bar', color='purple')
plt.title("Average Engagement Rate by Category (%)")
plt.ylabel("Engagement %")
plt.xlabel("Category")
plt.show()

```
